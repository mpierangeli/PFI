```python
import matplotlib.pyplot as plt
import numpy as np

# Read data from the text file
with open('history_unet3plus_relu_he_dicebce.txt', 'r') as file:
    data_str = file.read()

# Convert the string representation of dictionary to actual dictionary
data = eval(data_str)
# Increase the plot size
plt.figure(figsize=(10, 8))

# Create a figure and axis for the plot with larger size
fig, ax1 = plt.subplots(figsize=(10, 6))

# Plot the Dice curve on the primary y-axis
color = 'tab:red'
ax1.set_xlabel('Epochs')
ax1.set_ylabel('Dice Score', color=color)
ax1.plot(data['Dice'], color=color)
ax1.tick_params(axis='y', labelcolor=color)

# Set y-axis ticks with additional ticks from 0.8 to 1
yticks = np.arange(0, 1.1, 0.1)
ax1.set_yticks(yticks)

# Add lines indicating the maximum value on the Dice axis
max_index_dice = np.argmax(data['Dice'])
max_value_dice = data['Dice'][max_index_dice]

ax1.axvline(x=max_index_dice, color='red', linestyle='--', linewidth=0.8)
ax1.axhline(y=max_value_dice, color='blue', linestyle='--', linewidth=0.8)

# Create a secondary y-axis for the Loss curve
ax2 = ax1.twinx()
color = 'tab:blue'
ax2.set_ylabel('Loss', color=color)
ax2.plot(data['loss'], color=color)
ax2.tick_params(axis='y', labelcolor=color)

# Add lines indicating the minimum value on the Loss axis
min_index_loss = np.argmin(data['loss'])
min_value_loss = data['loss'][min_index_loss]

ax2.axvline(x=min_index_loss, color='green', linestyle='--', linewidth=0.8)
ax2.axhline(y=min_value_loss, color='purple', linestyle='--', linewidth=0.8)

# Add horizontal grid lines
ax1.grid(axis='y', linestyle='--', alpha=0.7, which='both')
ax2.grid(axis='y', linestyle='--', alpha=0.7, which='both')

# Display indicators for Max Dice and Min Loss inside the plot
ax1.annotate(f'Max Dice: {max_value_dice:.3f}', xy=(max_index_dice, max_value_dice),
             xytext=(max_index_dice - 3, max_value_dice - 0.05),
             fontsize=8, color='blue')
ax2.annotate(f'Min Loss: {min_value_loss:.3f}', xy=(min_index_loss, min_value_loss),
             xytext=(min_index_loss - 3, min_value_loss + 0.05),
             fontsize=8, color='purple')

# Display the legend at the middle right
fig.tight_layout()

plt.title('UNet3+ (RELU)')
plt.show()
```

```python
with open('dices_by_th.txt', 'r') as file:
    data_str = file.read()

all_curves = eval(data_str)

fig, ax = plt.subplots(figsize=(12, 10))

model_names = ['Model A', 'Model B', 'Model C', 'Model D', 'Model E', 'Model F', 'Model G', 'Model H', 'Model I']

for i, curve in enumerate(all_curves):
    threshold_values = np.linspace(0.05, 0.95, len(curve) - 1)
    ax.plot(threshold_values, curve[:-1], label=model_names[i])

    max_threshold = threshold_values[np.argmax(curve[:-1])]
    max_dice_value = np.max(curve[:-1])

    ax.axvline(x=max_threshold, linestyle='--', color=f'C{i}', alpha=0.7)
    #ax.axhline(y=max_dice_value, linestyle='--', color=f'C{i}', alpha=0.7)

    #ax.text(max_threshold, max_dice_value, f'{model_names[i]}\n{max_dice_value:.3f}', va='bottom', ha='center', color=f'C{i}')

yticks = np.arange(0.73, 0.86, 0.01)
ax.set_yticks(yticks)

xticks = np.arange(0.05, 1.0, 0.1)
ax.set_xticks(xticks)

ax.grid(axis='both', linestyle='--', alpha=0.7, which='both')

ax.legend(loc='upper left')

ax.set_xlabel('Threshold')
ax.set_ylabel('Dice Score')
plt.title('Dice Scores vs. Threshold for Multiple Models')

plt.show()
```

```python
with open('dices_by_th.txt', 'r') as file:
    data_str = file.read()

all_curves = eval(data_str)

fig, ax = plt.subplots(figsize=(14, 10))

model_names = ["AttUnet w/s","Unet3+","vsUnet","sUnet","Unet","Unet w/s","ResAttUnet","ClassicUnet","AttUnet"]

for i, curve in enumerate(all_curves):
    threshold_values = np.linspace(0.05, 0.95, len(curve) - 1)
    max_threshold = threshold_values[np.argmax(curve[:-1])]
    max_dice_value = np.max(curve[:-1])

    if 0.78 <= max_dice_value <= 0.85:
        ax.plot(threshold_values, curve[:-1], label=model_names[i])
        ax.axvline(x=max_threshold, linestyle='--', color=f'C{i}', alpha=0.7)
        #ax.axhline(y=max_dice_value, linestyle='--', color=f'C{i}', alpha=0.7)
        #ax.text(max_threshold, max_dice_value, f'{model_names[i]}\n{max_dice_value:.3f}', va='bottom', ha='center', color=f'C{i}')

yticks = np.arange(0.78, 0.88, 0.01)
ax.set_yticks(yticks)

xticks = np.arange(0.05, 1.0, 0.1)
ax.set_xticks(xticks)

ax.grid(axis='both', linestyle='--', alpha=0.7, which='both')

ax.legend(loc='upper left')

ax.set_xlabel('Threshold')
ax.set_ylabel('Dice Score')
ax.set_ylim(0.78, 0.85)  # Set y-axis limits

plt.title('Dice Scores vs. Threshold for Multiple Models')

plt.show()
```
